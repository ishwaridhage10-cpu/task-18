```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# 1. Load Dataset
df = pd.read_csv("/content/ford_car_dataset.csv")

# Check Missing Values
print("Missing Values:")
print(df.isnull().sum())

# Remove Duplicate Rows
df = df.drop_duplicates()

print("\nShape After Cleaning:")
print(df.shape)

# 2. Perform EDA

# Histogram
for c in ['price', 'mileage', 'year', 'engineSize', 'mpg']:
    plt.figure(figsize=(6,4))
    sns.histplot(df[c], bins=25)
    plt.title(c)
    plt.show()

# Count Plot
for c in ['fuelType', 'transmission', 'model']:
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x=c)
    plt.title(c)
    plt.xticks(rotation=45)
    plt.show()

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# 3. Identify Input and Output Features
X = df.drop("price", axis=1)
y = df["price"]

print("\nIndependent Features:")
print(X.columns)

print("\nDependent Feature:")
print(y.name)

# 4. One-Hot Encoding
encoded_df = pd.get_dummies(df, columns=['fuelType', 'transmission'])

print("\nEncoded Dataset Shape:")
print(encoded_df.shape)

# 5. Feature Scaling
scaler = StandardScaler()

numeric_features = ['year', 'mileage', 'tax', 'mpg', 'engineSize']

encoded_df[numeric_features] = scaler.fit_transform(encoded_df[numeric_features])

print("\nFirst 5 Rows of Scaled Data:")
print(encoded_df.head())
```
